```python
import pandas as pd
```

```python
usd_cny_df = pd.read_csv('./Data/USD_CNY Historical Data.csv', index_col = 'Date', parse_dates = True, infer_datetime_format = True )
```

```python
usd_cny_df.head()
```

```python
usd_cny_df= usd_cny_df.sort_values(by = 'Date')
```

```python
usd_cny_df.tail()
```

```python
usd_cny_df.isna().sum()
```

```python
len(usd_cny_df)
```

```python
usd_cny_df = usd_cny_df.drop(columns = ['Open', 'High', 'Low'])
```

```python
usd_cny_df = usd_cny_df.rename(columns = {'Price': 'USD/CNY_Price', 'Change %': 'USD/CNY_PCT_CHANGE'})
```

```python
usd_cny_df.head()
```

```python
usd_eur_df = pd.read_csv('./Data/USD_EUR Historical Data.csv', index_col = 'Date', parse_dates = True, infer_datetime_format = True)
```

```python
usd_eur_df.head()
```

```python
usd_eur_df = usd_eur_df.sort_values(by = 'Date')
```

```python
usd_eur_df.tail()
```

```python
usd_eur_df.isna().sum()
```

```python
len(usd_eur_df)
```

```python
usd_eur_df = usd_eur_df.drop(columns = ['Open', 'High', 'Low'])
```

```python
usd_eur_df= usd_eur_df.rename(columns = {'Price': 'USD/EUR_Price', 'Change %': 'USD/EUR_PCT_CHANGE'})
```

```python
usd_eur_df.head()
```

```python
usd_jpy_df = pd.read_csv('./Data/USD_JPY Historical Data.csv', index_col = 'Date', parse_dates = True, infer_datetime_format = True)
```

```python
usd_jpy_df.head()
```

```python
usd_jpy_df = usd_jpy_df.sort_values(by = 'Date')
```

```python
usd_jpy_df.tail()
```

```python
usd_jpy_df.isna().sum()
```

```python
len(usd_jpy_df)
```

```python
usd_jpy_df = usd_jpy_df.drop(columns = ['Open', 'High', 'Low'])
```

```python
usd_jpy_df= usd_jpy_df.rename(columns = {'Price': 'USD/JPY_Price', 'Change %': 'USD/JPY_PCT_CHANGE'})
```

```python
usd_jpy_df.head()
```

```python
currency_list = [usd_cny_df, usd_eur_df, usd_jpy_df]
```

```python
currency_df = pd.concat(currency_list, axis = 1)
```

```python
currency_df.tail()
```

```python
btc_df = pd.read_csv('./Data/BTC_daily_price.csv', index_col = 'Date', parse_dates = True, infer_datetime_format = True)
```

```python
btc_df.head()
```

```python
btc_df = btc_df.drop(columns = {'Open', 'High', 'Low', 'Close'})
```

```python
btc_df.head()
```

```python
btc_df = btc_df.rename(columns = {'Adj Close': 'BTC_Price', 'Volume': 'BTC_Volume'})
```

```python
btc_df.head()
```

```python
final_currency_btc_df = pd.concat([btc_df, currency_df], axis = 1)
```

```python
final_currency_btc_df.head()
```

```python
final_currency_btc_df['USD/CNY_Price'] = final_currency_btc_df['USD/CNY_Price'].ffill()
```

```python
final_currency_btc_df['USD/EUR_Price'] = final_currency_btc_df['USD/EUR_Price'].ffill()
```

```python
final_currency_btc_df['USD/JPY_Price'] = final_currency_btc_df['USD/JPY_Price'].ffill()
```

```python
final_currency_btc_df.head()
```

```python
final_currency_btc_df['USD/CNY_PCT_CHANGE'] = final_currency_btc_df['USD/CNY_PCT_CHANGE'].str.replace(r'\%', '').astype(float)
```

```python
final_currency_btc_df.head()
```

```python
final_currency_btc_df['USD/EUR_PCT_CHANGE'] = final_currency_btc_df['USD/EUR_PCT_CHANGE'].str.replace(r'\%', '').astype(float)
```

```python
final_currency_btc_df.head()
```

```python
final_currency_btc_df['USD/JPY_PCT_CHANGE'] = final_currency_btc_df['USD/JPY_PCT_CHANGE'].str.replace(r'\%', '').astype(float)
```

```python
final_currency_btc_df.head()
```

```python
final_currency_btc_df = final_currency_btc_df.fillna(0)
```

```python
final_currency_btc_df.head()
```

```python
final_currency_btc_df.to_csv('./Data/final_currency_df.csv')
```
